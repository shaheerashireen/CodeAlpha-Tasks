```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
np.random.seed(42)

data = {
    "Student_ID": range(1, 31),
    "Gender": np.random.choice(["Male", "Female"], 30),
    "Study_Hours": np.random.randint(1, 10, 30),
    "Attendance_Percentage": np.random.randint(60, 100, 30),
    "Midterm_Score": np.random.randint(40, 85, 30),
    "Final_Score": np.random.randint(45, 90, 30)
}

df = pd.DataFrame(data)
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
avg_attendance = df.groupby("Attendance_Percentage")["Final_Score"].mean()

plt.plot(avg_attendance.index, avg_attendance.values)
plt.xlabel("Attendance Percentage")
plt.ylabel("Average Final Score")
plt.title("Attendance vs Average Final Score")
plt.show()
```

```python
avg_scores = df.groupby("Study_Hours")["Final_Score"].mean()

plt.plot(avg_scores.index, avg_scores.values)
plt.xlabel("Study Hours")
plt.ylabel("Average Final Score")
plt.title("Study Hours vs Average Final Score")
plt.show()
```

```python
gender_avg = df.groupby("Gender")["Final_Score"].mean()

gender_avg.plot(kind="bar")
plt.xlabel("Gender")
plt.ylabel("Average Final Score")
plt.title("Average Final Score by Gender")
plt.show()
```

```python
import seaborn as sns

sns.boxplot(data=df[["Midterm_Score", "Final_Score"]])
plt.title("Comparison of Midterm and Final Scores")
plt.show()
```
